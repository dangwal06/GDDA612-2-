```python
# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["GD612"]  
collection = db["Assesment_2"] 

# Count the number of documents in the collection
document_count = collection.count_documents({})
print("Number of documents in the collection:", document_count)
```

```python
# Group documents by brand and calculate the average price for each brand
pipeline = [
    {
        "$group": {
            "_id": "$brand",
            "average_price": {"$avg": "$price"},
            "total_quantity_sold": {"$sum": "$quantity_sold"}
        }
    },
    {
        "$sort": {"average_price": -1}  # Sort by average_price in descending order
    }
]

result = collection.aggregate(pipeline)

# Display the result
for doc in result:
    print(doc)
```

```python
# Updating single document 
result = collection.update_one({"brand": "OEM"}, {"$set": {"price": 10.0}})

# Print the result
print("Matched:", result.matched_count)
print("Modified:", result.modified_count)
```

```python
# Updating all documents
result = collection.update_many({"brand": "OEM"}, {"$set": {"price": 10.0}})

# Print the result
print("Matched:", result.matched_count)
print("Modified:", result.modified_count)
```
